```python
import pandas as pd
from pycaret.datasets import get_data
from pycaret.classification import *

data = pd.read_csv('C:/Users/ekyfe/Documents/college/Semester 4/MLOps/praktikum/stroke/healthcare-dataset-stroke-data.csv')

# Drop the 'id' column
data.drop(columns=['id'], inplace=True)

# Fill missing values with mean
data.fillna(data.mean(), inplace=True)

# Setup PyCaret
s = setup(data, target='stroke', session_id=123)

# Compare models
best = compare_models()

# Tune best model
model = tune_model(best)

# Evaluate model
evaluate_model(model)

# Finalize model
finalize_model(model)

# Save model
save_model(model, 'stroke_pipeline')
```
